# Introduction to plotting
Python is a great tool for plotting functions. Any equation in a physics textbook can be plotted somehow. It is a good skill to learn how to plot functions. This notebook takes you through a template for plotting and customizing your plot to look nice.

```python
# import relevant packages
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

```python
# define constants
mu0 = 1.257e-6     # N/A^2 or T⋅m/A    https://en.wikipedia.org/wiki/Vacuum_permeability    
eps0 = 8.854e-12   # F/m               https://en.wikipedia.org/wiki/Vacuum_permittivity
c_light = 3e8      # m/s               https://en.wikipedia.org/wiki/Speed_of_light
h=6.626e-34        # J⋅s               https://en.wikipedia.org/wiki/Planck_constant
hbar = 1.0546e-34  # J⋅s               https://en.wikipedia.org/wiki/Planck_constant
m_e = 9.1e-31      # kg                https://en.wikipedia.org/wiki/Electron_rest_mass
q_e = 1.6e-19      # C                 https://en.wikipedia.org/wiki/Elementary_charge
k_B_J = 1.38e-23     # J/K               https://en.wikipedia.org/wiki/Boltzmann_constant
k_B_eV =8.617333262145e-5  # eV/K
```

```python
# define variables
x_values =  np.linspace(-1, 1, 1001)

I0 = 1e-10 # Amps
Temp = 300 # Kelvin

y_values = x_values**2
```

```python
x_values
```

```python
y_values = x_values**3
```

```python
#############################################
### begin simple plotting ###################
#############################################

plt.plot(x_values,y_values)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Plot of function')
plt.grid(True)
plt.show()
```

```python
#############################################
### more involved plotting ##################
#############################################

# define some function
y_values = np.sin(2*np.pi*x_values)

plt.figure(figsize=(8,6))
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
plt.subplot(211)
# make a simple plot
plt.plot(x_values,y_values)
plt.xlabel('x axis',fontsize=20)
plt.ylabel('y axis',fontsize=20)
plt.title('Sinusoidal function',fontsize=25)
plt.grid(True)


# define some other functions -- note how we can call special functions like sin and exp from the numpy library
# https://docs.scipy.org/doc/numpy/reference/routines.math.html
y_values1 = np.exp( np.sin(2*np.pi*x_values) )
y_values2 = np.sin( 2*np.pi* np.exp(x_values) )

# plot two curves in the same graph -- note how we can make labels and change the range that is plotted:
plt.subplot(212)
plt.plot(x_values,y_values1,'-.r',label='red values')
plt.plot(x_values,y_values2,'--g',label='green values')
plt.xlabel('Voltage (V)',fontsize=20)
plt.ylabel('current (A)',fontsize=20)
plt.title('Comparing interesting curves',fontsize=25)
plt.grid(True)
plt.legend(frameon=False,loc='best',fontsize = 15) 
plt.ylim([-2,5])


plt.show()

# optional: save as a png file
# plt.savefig("fxn.png",dpi=400)
```
